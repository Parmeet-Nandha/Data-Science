Google daily Serach Analysis 

 Import the necessary Python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# we have to install AN api which is provided by Google to analyse the daily serach on Google
# Install the api by using pip command
!pip install pytrends
# import TrendReq from pytrends.request
from pytrends.request import TrendReq
#it will allows a interface for automating downloading the reports of Google
trends = TrendReq()
```

```python
# Generate List by keyword Data Analysis
trends.build_payload(kw_list=["Data Analysis"])
#popular search by different region
data = trends.interest_by_region()
# sort the data
data = data.sort_values(by="Data Analysis", ascending=False)
#print first 10 rows of popular search
data = data.head(10)
print(data)
```

```python
#plot the bar graph using matplot
data.reset_index().plot(x="geoName", y="Data Analysis", 
                        figsize=(20,15), kind="bar")
# we can use diffent style such as 538, ggplot,grayscale
plt.style.use('fivethirtyeight')
plt.show()
```

# Plot the total google search for keyword for a time period

```python
data = TrendReq(hl='en-AU', tz=360)
data.build_payload(kw_list=['Data Analysis'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Data Analysis'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Analysis', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

It is clearly seen from the line graph that google search for keyword data analysis is increased.
